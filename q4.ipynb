```python
# Import libraries
import pandas as pd
```

```python
# Load dataset to variable data
data = pd.read_csv("wind.data", sep=" ")
data.head()
```

```python
# Add date column
data["date"] = data["Dy"].astype(str) + "-" + data["Mo"].astype(str) + "-" + data["Yr"].astype(str)
data["date"] = pd.to_datetime(data["date"]).dt.date
data.head()
```

```python
# Sum of missing values
data.isna().sum()
```

```python
# Sum of non-missing values
data.count()
```

```python
# Mean over all locations
temp = data.drop(labels=["Yr", "Mo", "Dy", "date"], axis=1)
temp = temp.fillna(0)
temp.to_numpy().mean()
```

```python
# Min, max, mean, sd
loc_stats = data
print(f"Min: {loc_stats.min()}")
print(f"Max: {loc_stats.max()}")
print(f"Mean: {loc_stats.mean()}")
print(f"SD: {loc_stats.std()}")
```
